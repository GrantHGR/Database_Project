```python
%%sql # creates table displaying each store and its number of employees
SELECT store_id AS Store, COUNT(*) AS 'Number of Employees'
FROM Employee
GROUP BY store_id
```

```python
%%sql # queries full names and email addresses for easily creating company wide contact information
SELECT fullname AS 'Full Name', phone_num AS 'Phone Number', email_address AS 'Email Address'
FROM Employee
```

```python
%%sql # query using INNER JOIN to create full list of employees with calculation for yearly income using hourly pay multiplied by weekly hours and ordered by gross yearly income DESC
SELECT DISTINCT Employee.emp_id AS 'Employee ID', Employee.fullname AS 'Full Name', Pay.pay_rate AS 'Hourly Pay', Employee.weekly_hours AS 'Hours per Week', (52*(Pay.pay_rate*Employee.weekly_hours)) AS 'Gross Yearly Income'
FROM Employee
INNER JOIN Pay ON Employee.emp_id=Pay.emp_id
ORDER BY (52*(Pay.pay_rate*Employee.weekly_hours)) DESC;
```

```python
%%sql # USING 3 WAY JOIN displays names and positions of employees grouped by store 
SELECT Employee.fullname AS 'Full Name', Store.store_id AS 'Store', Position.position_name AS 'Position'
FROM Employee
INNER JOIN Store ON Employee.store_id=Store.store_id
INNER JOIN Position ON Employee.position_id = Position.position_id
GROUP BY Employee.fullname, Store.store_id, Position.position_name;
```

```python
%%sql # FUNCTION produces the total cost of business to employ its workers 
SELECT SUM((52*(Pay.pay_rate*Employee.weekly_hours))) AS 'Total Yearly Cost to Business'
FROM Employee
INNER JOIN Pay ON Employee.emp_id=Pay.emp_id
ORDER BY (52*(Pay.pay_rate*Employee.weekly_hours)) DESC;
```

```python
%%sql
SELECT store_id AS 'Store Number', SUM(((Employee.weekly_hours))) AS 'Total Hours Per Week'
FROM Employee
INNER JOIN Pay ON Employee.emp_id=Pay.emp_id
GROUP BY store_id;
```
